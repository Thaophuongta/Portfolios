```python
# Q1 = 25th = 100
# Q3 = 75th = 110
# IQR = Q3 - Q1 = 10
# LW = Q1 - 1.5*IQR = 85 --> there are outliers on the left (less than 85)
# RW = Q3 + 1.5*IQR = 125 --> there are outliers on the right (over than 125)
```

```python
excloutlierscalories=data[(data['calories']>=85) & (data['calories']<=125)]
excloutlierscalories # Subset that remove the outsiders for calories factor
```
